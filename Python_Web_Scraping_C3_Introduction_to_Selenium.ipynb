```python
import sys
import os
import time

from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager

import pandas as pd
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())

base_url = 'https://merchant.thaichana.com/search?'
para = 'size=10&page=0&category=ทุกประเภท&province=กรุงเทพมหานคร&rating=0&passQuestionnaire=false&lockLatLong=false'

url = base_url + para

driver.get(url)
```

```python
thaichana_data = pd.DataFrame([])

main_list = driver.find_elements_by_class_name('css-1rrtwvw')[0]

for item in main_list.find_elements_by_tag_name('a'):
    if item is None:
        continue
    
    item_url = item.get_attribute('href')
    item_name = item.find_elements_by_class_name('css-11jnqim')
    item_address = item.find_elements_by_class_name('ant-typography')
    item_type = item.find_elements_by_class_name('css-72rvq0')
    item_rating = item.find_elements_by_class_name('css-1uv43ib')
    
    if not item_name:
        continue
    else:
        item_name = item_name[0].text
    
    item_address = item_address[0].text if item_address else ''
    item_type = item_type[0].text if item_type else ''
    item_rating = item_rating[0].text if item_rating else ''
        
    thaichana_data = thaichana_data.append(pd.DataFrame({
      'Name': item_name,
      'Type': item_type,
      'Rating': item_rating,
      'Address': item_address,
      'internal_url': item_url,
      }, index=[0]), ignore_index=True)

```

```python
MAX_PAGE = 3

counter = 1
thaichana_data = pd.DataFrame([])


while counter <= MAX_PAGE:
    print('Page: ', counter)
    
    main_list = driver.find_elements_by_class_name('css-1rrtwvw')[0]
    
    for item in main_list.find_elements_by_tag_name('a'):
        if item is None:
            continue

        item_url = item.get_attribute('href')
        item_name = item.find_elements_by_class_name('css-11jnqim')
        item_address = item.find_elements_by_class_name('ant-typography')
        item_type = item.find_elements_by_class_name('css-72rvq0')
        item_rating = item.find_elements_by_class_name('css-1uv43ib')

        if not item_name:
            continue
        else:
            item_name = item_name[0].text

        item_address = item_address[0].text if item_address else ''
        item_type = item_type[0].text if item_type else ''
        item_rating = item_rating[0].text if item_rating else ''

        thaichana_data = thaichana_data.append(pd.DataFrame({
          'Name': item_name,
          'Type': item_type,
          'Rating': item_rating,
          'Address': item_address,
          'internal_url': item_url,
          }, index=[0]), ignore_index=True)
    
    # Go to next page.
    nextpage_link = driver.find_elements_by_class_name("css-tsdlxr")[0].find_elements_by_tag_name('a')[0]
    nextpage_link.click()
    
    counter = counter + 1
    time.sleep(2)
    
driver.close()
```
